```python
def drop_outliers(df, field_name):
    iqr = 1.5 * (np.percentile(df_new[field_name], 75) - np.percentile(df_new[field_name], 25))
    df_new.drop(df_new[df_new[field_name] > (iqr + np.percentile(df_new[field_name], 75))].index, inplace=True)
    df_new.drop(df_new[df_new[field_name] < (np.percentile(df_new[field_name], 25) - iqr)].index, inplace=True)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score, log_loss, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Assume X and Y are defined
# X, Y = ...

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=10)

models = [
    ('LogisticRegression', LogisticRegression()),
    ('KNN', KNeighborsClassifier()),
    ('SVM', SVC(probability=True)),  # SVC needs probability=True for ROC AUC
    ('Decision Tree Classifier', DecisionTreeClassifier()),
    ('Decision Tree Classifier(gini)', DecisionTreeClassifier(criterion='gini', max_depth=5)),
    ('Decision Tree Classifier(entropy)', DecisionTreeClassifier(criterion='entropy', max_depth=5)),
    ('Random Forest Classifier', RandomForestClassifier(max_depth=7)),  # Adjusted max_depth to a valid integer
    ('XGB', XGBClassifier(eta=0.01, gamma=10, reg_lambda=4))
]

for name, model in models:
    model.fit(x_train, y_train)

    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)

    bag = BaggingClassifier(
        base_estimator=model,
        n_estimators=500,
        max_samples=100,
        bootstrap=True,
        random_state=42
    )

    bag.fit(x_train, y_train)
    y_pred_train_bag = bag.predict(x_train)
    y_pred_test_bag = bag.predict(x_test)

    bag_dt_train = accuracy_score(y_train, y_pred_train_bag)
    bag_dt_test = accuracy_score(y_test, y_pred_test_bag)
    print("Base Estimator:", name)
    print("Bagging accuracy (Train):", (bag_dt_train * 100).round(2))
    print("Bagging accuracy (Test):", (bag_dt_test * 100).round(2))

    cm_train = confusion_matrix(y_train, y_pred_train_bag)
    print("Confusion Matrix (Train):\n", cm_train)
    print("Accuracy Score (Train):", (bag_dt_train * 100).round(2))
    
    TN, FP, FN, TP = cm_train.ravel()
    TNR = TN / (TN + FP)
    print("Specificity (Train):", (TNR * 100).round(2))

    rc = recall_score(y_train, y_pred_train_bag)
    print("Sensitivity (Train):", (rc * 100).round(2))

    pc = precision_score(y_train, y_pred_train_bag)
    print("Precision (Train):", (pc * 100).round(2))

    f1sc_train = f1_score(y_train, y_pred_train_bag)
    print("F1 Score (Train):", (f1sc_train * 100).round(2))

    logloss_train = log_loss(y_train, y_pred_train_bag)
    print("Log Loss (Train):", logloss_train.round(2))

    Y_predict_prob = bag.predict_proba(x_train)[:, 1]
    fpr, tpr, _ = roc_curve(y_train, Y_predict_prob)
    plt.plot(fpr, tpr)
    plt.ylabel("TPR - True positive rate")
    plt.xlabel("FPR - False positive rate")
    plt.title(f"ROC Curve (Train) - {name}")
    plt.show()

    auc = roc_auc_score(y_train, Y_predict_prob)
    print("Area under Curve (Train):", auc.round(4))

    cm_test = confusion_matrix(y_test, y_pred_test_bag)
    print("Confusion Matrix (Test):\n", cm_test)
    print("Accuracy Score (Test):", (bag_dt_test * 100).round(2))
    
    TN, FP, FN, TP = cm_test.ravel()
    TNR = TN / (TN + FP)
    print("Specificity (Test):", (TNR * 100).round(2))

    rc = recall_score(y_test, y_pred_test_bag)
    print("Sensitivity (Test):", (rc * 100).round(2))

    pc = precision_score(y_test, y_pred_test_bag)
    print("Precision (Test):", (pc * 100).round(2))

    f1sc_test = f1_score(y_test, y_pred_test_bag)
    print("F1 Score (Test):", (f1sc_test * 100).round(2))

    logloss_test = log_loss(y_test, y_pred_test_bag)
    print("Log Loss (Test):", logloss_test.round(2))

    Y_predict_prob_test = bag.predict_proba(x_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, Y_predict_prob_test)
    plt.plot(fpr, tpr)
    plt.ylabel("TPR - True positive rate")
    plt.xlabel("FPR - False positive rate")
    plt.title(f"ROC Curve (Test) - {name}")
    plt.show()

    auc = roc_auc_score(y_test, Y_predict_prob_test)
    print("Area under Curve (Test):", auc.round(4))
    print("-" * 80)
```
